```python
#Learning performance

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#test-score variance figure for different thresholds

data = pd.read_csv("learning_nth_6.csv")
data2 = pd.read_csv("learning_nth_4.csv")
data3 = pd.read_csv("learning_nth_8.csv")
#data3 = pd.read_csv("learning_nth_70.csv")

data.columns=["f1_score", "test_score", "variance"]
data2.columns=["f1_score", "test_score", "variance"]
data3.columns=["f1_score", "test_score", "variance"]

var1 = np.sqrt(data["variance"])
var2 = np.sqrt(data2["variance"])
var3 = np.sqrt(data3["variance"])




plt.plot(var1, data["f1_score"], "*")
plt.plot(var2, data2["f1_score"], "*")
plt.plot(var3, data3["f1_score"], "*")
#plt.legend(["nth=50", "nth=60", "nth=70"])
plt.xlabel("variance")
plt.ylabel("f1_score")
plt.show()

plt.plot(var1, data["test_score"], "*")
plt.plot(var2, data2["test_score"], "*")
plt.plot(var3, data3["test_score"], "*")
plt.legend(["nth=6", "nth=4", "nth=8"])
plt.xlabel("variance")
plt.ylabel("test accuracy")
plt.show()
```
